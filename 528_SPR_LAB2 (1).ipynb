```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to compute DTFT
def dtft(x, fs, n_points=1024):
    N = len(x)
    X = np.fft.fftshift(np.fft.fft(x, n_points))
    f = np.linspace(-fs/2, fs/2, n_points)
    return f, np.abs(X)/N
```

```python
# Function to compute DFT
def dft(x, fs):
    N = len(x)
    X = np.fft.fft(x)
    f = np.fft.fftfreq(N, 1/fs)
    return f, np.abs(X)/N
```

```python
# Sampling parameters
fs = 1000       # Sampling frequency
t = np.arange(0, 1, 1/fs)  # Time vector, 1 second duration
```

```python
# Q1: Single Sinusoid
# A smooth, repeating wave (sine or cosine).
# Natural phenomena like sound waves, AC electricity, or vibrations.

f1 = 5  # Frequency 5 Hz
x1 = np.sin(2*np.pi*f1*t)

plt.figure(figsize=(12,4))
plt.plot(t, x1)
plt.title("Q1: Time-domain Signal (5 Hz Sine)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

# DTFT
f_dtft, X_dtft = dtft(x1, fs)
plt.figure(figsize=(12,4))
plt.plot(f_dtft, X_dtft)
plt.title("Q1: DTFT of 5 Hz Sine")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()

# DFT
f_dft, X_dft = dft(x1, fs)
plt.figure(figsize=(12,4))
plt.stem(f_dft, X_dft, use_line_collection=True)
plt.title("Q1: DFT of 5 Hz Sine")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()

# Inference:
# - Time-domain waveform shows a smooth sinusoid at 5 Hz.
# - DTFT and DFT show a single peak at 5 Hz, representing the signal frequency.
```

```python
# Q2: Composite Signal
#Combination of two or more sine waves with different frequencies and amplitudes.
# Music, speech, any real-world signal with multiple components.

f2 = 10
x2 = 0.7*np.sin(2*np.pi*f1*t) + 0.5*np.sin(2*np.pi*f2*t)

plt.figure(figsize=(12,4))
plt.plot(t, x2)
plt.title("Q2: Time-domain Composite Signal")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

# DTFT
f_dtft, X_dtft = dtft(x2, fs)
plt.figure(figsize=(12,4))
plt.plot(f_dtft, X_dtft)
plt.title("Q2: DTFT of Composite Signal")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()

# DFT
f_dft, X_dft = dft(x2, fs)
plt.figure(figsize=(12,4))
plt.stem(f_dft, X_dft, use_line_collection=True)
plt.title("Q2: DFT of Composite Signal")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
# Inference:
# - Time-domain waveform shows combined oscillations.
# - Frequency-domain shows two distinct peaks at 5 Hz and 20 Hz, corresponding to the individual sinusoids.
```

```python
# Q3: Exponentially Decaying Signal
#Starts with high amplitude and gradually decreases to zero over time
#(Damped vibrations, capacitor discharges, radioactive decay)
tau = 0.2
x3 = np.exp(-t/tau)

plt.figure(figsize=(12,4))
plt.plot(t, x3)
plt.title("Q3: Time-domain Exponentially Decaying Signal")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

# DTFT
f_dtft, X_dtft = dtft(x3, fs)
plt.figure(figsize=(12,4))
plt.plot(f_dtft, X_dtft)
plt.title("Q3: DTFT of Exponentially Decaying Signal")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()

# DFT
f_dft, X_dft = dft(x3, fs)
plt.figure(figsize=(12,4))
plt.stem(f_dft, X_dft, use_line_collection=True)
plt.title("Q3: DFT of Exponentially Decaying Signal")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```

```python
# Q4: Rectangular Pulse
# Constant value for a short period, then zero (Digital signals, radar pulses, switching signals).

width = 0.1  # 100 ms pulse
x4 = np.where((t >= 0.3) & (t <= 0.3+width), 1, 0)

plt.figure(figsize=(12,4))
plt.plot(t, x4)
plt.title("Q4: Time-domain Rectangular Pulse")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

# DTFT
f_dtft, X_dtft = dtft(x4, fs)
plt.figure(figsize=(12,4))
plt.plot(f_dtft, X_dtft)
plt.title("Q4: DTFT of Rectangular Pulse")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()

# DFT
f_dft, X_dft = dft(x4, fs)
plt.figure(figsize=(12,4))
plt.stem(f_dft, X_dft, use_line_collection=True)
plt.title("Q4: DFT of Rectangular Pulse")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```

```python
# We have signals (sine, composite, exponential, rectangular) in the time domain  how they look over time.
# We want to see what frequencies make up the signal  frequency domain.
# To do this, we use DTFT (Discrete-Time Fourier Transform) and DFT (Discrete Fourier Transform).

DTFT(Discrete-Time Fourier Transform): Shows the ideal, continuous spectrum of a signal.
DFT(Discrete Fourier Transform): Shows a computable, discrete spectrum  what we can calculate on a computer.
Comparing them helps us see how the continuous ideal spectrum (DTFT) is represented in practical digital computations (DFT).
```

```python
| Signal Type       | Time Domain                   | Frequency Domain (DTFT vs DFT) | What to Observe                                             |
| ----------------- | ----------------------------- | ------------------------------ | ----------------------------------------------------------- |
| Sine              | Smooth periodic wave          | Single peak at sine frequency  | DFT matches DTFT at discrete points                         |
| Composite         | Sum of 2+ sine waves          | Peaks at each frequency        | Peaks show all components; DFT approximates DTFT            |
| Exponential Decay | Amplitude decreases over time | Broad frequency range          | Time decay  spread in frequency; DFT shows sampled version |
| Rectangular Pulse | Short duration pulse          | Sinc-shaped spectrum           | Short in time  wide frequency  DFT shows sampled sinc     |
```

```python
| Feature    | **DTFT**                              | **DFT**                              |
| ---------- | ------------------------------------- | ------------------------------------ |
| Domain     | Continuous frequency                  | Discrete frequency (samples of DTFT) |
| Input      | Discrete-time signal                  | Discrete-time signal (finite length) |
| Output     | Continuous spectrum (infinite points) | Finite number of frequency bins      |
| Usage      | Theoretical analysis                  | Practical computation using FFT      |
| Resolution | Infinite frequency resolution         | Limited by number of points (N)      |
```
